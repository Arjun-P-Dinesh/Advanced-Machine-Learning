```python
model1=y.mean()
model2=1+4.3*x
model3=2+3.2*x
data['model1']=model1
data['model2']=model2
data['model3']=model3
```

```python
fig, ax=plt.subplots()
#plotting actual minutes
ax.scatter(x='Avg_rooms ',y='MValue',data=data,label='Median Value per room size')
ax.plot(data['Avg_rooms '],data['model1'],color='red',label='Model_1')

ax.plot(data['Avg_rooms '],data['model2'],color='blue',label='Model_2')

ax.plot(data['Avg_rooms '],data['model3'],color='purple',label='Model_3')

ax.set_ylabel('Median Home Value')
ax.set_xlabel('Average No of Rooms')
ax.set_title('Speculated Models')
ax.legend()
```

```python
model1_obs=pd.DataFrame({
    'Average No Rooms':data['Avg_rooms '],
    'Actual':data['Avg_rooms '],
    'Predicted':data['model1'],
    'Error':data['model1']-data['Avg_rooms ']
})
model1_obs
```

```python

model2_obs=pd.DataFrame({
    'Average No Rooms':data['Avg_rooms '],
    'Actual':data['Avg_rooms '],
    'Predicted':data['model2'],
    'Error':data  ['model2']-data['Avg_rooms ']
})
model2_obs
```

```python
model3_obs=pd.DataFrame({
    'Average No Rooms':data['Avg_rooms '],
    'Actual':data['Avg_rooms '],
    'Predicted':data['model3'],
    'Error':data['model3']-data['Avg_rooms ']
})
model3_obs
```

```python
X=data['Avg_rooms '].values
y=data  ['MValue'].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
model=LinearRegression()
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
len(y_pred)
len(y_train)
```

```python
coefficient = model.coef_[0][0]
intercept = model.intercept_[0]
```

```python
print(f'Coefficient:{coefficient}')
print(f'Intercept:{intercept}')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_train)
mse_train = mean_squared_error(y_train, y_pred)
r2_train = r2_score(y_train, y_pred)

print(f'Training Set Metrics:')
print(f'Mean Squared Error: {mse_train}')
print(f'R-squared: {r2_train}')

y_pred=model.predict(X_test)
mse_test = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)

print('\nTesting Set Metrics:')
print(f'Mean Squared Error: {mse_test}')
print(f'R-squared: {r2_test}')
```
